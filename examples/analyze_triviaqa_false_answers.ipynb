```python
import requests
from bs4 import BeautifulSoup

def get_related_entities(entity):
    # Replace spaces with underscores for URL
    entity = entity.replace(' ', '_')
    
    # Wikipedia URL
    url = f'https://en.wikipedia.org/wiki/{entity}'
    
    # Get the HTML content
    response = requests.get(url)
    soup = BeautifulSoup(response.content, 'html.parser')
    
    # Find all links on the page
    links = soup.find_all('a')
    
    related_entities = set()

    for link in links:
        href = link.get('href', '')
        
        # Check if it's a link to another Wikipedia article
        if href.startswith('/wiki/') and not ':' in href:
            related_entity = href.split('/wiki/')[1].replace('_', ' ')
            related_entities.add(related_entity)

    return related_entities

# Example usage
related_entities = get_related_entities('Sinclair Lewis')
print(related_entities)
```
