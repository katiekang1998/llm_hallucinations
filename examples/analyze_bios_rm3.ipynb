```python
import numpy as np
import json
import matplotlib.pyplot as plt
```

```python
model_path = "ckpts/rm3_bios_llama7B/checkpoint_30000/hf_model"
num_total_preds_True = np.load(model_path+"/test_medium_True_num_total_preds.npy")
num_total_preds_False = np.load(model_path+"/test_medium_False_num_total_preds.npy")

num_correct_preds_True = np.load(model_path+"/test_medium_True_num_correct_preds.npy")
num_correct_preds_False = np.load(model_path+"/test_medium_False_num_correct_preds.npy")
```

```python
plt.hist(num_total_preds_True, bins=np.arange(0, 8, 1), alpha=0.5, label='True')
plt.hist(num_total_preds_False, bins=np.arange(0, 8, 1), alpha=0.5, label='False')
plt.legend(loc='upper right')
plt.legend()
```

```python
plt.hist(num_correct_preds_True, bins=np.arange(0, 8, 1), alpha=0.5, label='True')
plt.hist(num_correct_preds_False, bins=np.arange(0, 8, 1), alpha=0.5, label='False')
plt.legend(loc='upper right')
plt.legend()
```

```python
plt.hist(num_correct_preds_True/num_total_preds_True, bins=np.arange(0, 1, 0.2), alpha=0.5, label='True')
plt.hist(num_correct_preds_False/num_total_preds_False, bins=np.arange(0, 1, 0.2), alpha=0.5, label='False')
plt.legend(loc='upper right')
plt.legend()
```
